```python
# 데이터타입을 변환해서 사용할 수 있도록 datetime 형태로 바꾼다.
df['Month'] = pd.to_datetime(df['Month'])
df.dtypes
```

```python
indexedData = df.set_index('Month')
indexedData.head()
```

```python
indexedData.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomp = seasonal_decompose(indexedData)
decomp.plot()
```
