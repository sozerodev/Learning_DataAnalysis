```python
import pandas as pd
from plotnine import *
```

```python
pokemon = pd.read_csv('data/Pokemon.csv', index_col=0)
pokemon.shape
```

```python
# 공격과 방어에 대한 산점도
(ggplot(pokemon)
 + aes('Attack', 'Defense')
 + geom_point()
)
```

```python
(ggplot(pokemon)
 + aes('Attack','Defense', color='Legendary')
 + geom_point()
 + ggtitle('Legendary 상태에 따른 공격과 방어 산점도')
 + theme(text=element_text(family='NanumBarunGothic'))
)
```

```python
(ggplot(pokemon)
 + aes('HP', 'Attack', color='Legendary')
 + geom_point()
 + ggtitle('Legendary 상태에 따른 HP와 공격 산점도')
 + theme(text=element_text(family='NanumBarunGothic'))
)
```

```python
(ggplot(pokemon)
 + aes('Attack', 'Defense', color='HP')
 + geom_point()
 + ggtitle('HP에 따른 공격과 방어 산점도')
 + theme(text=element_text(family='NanumBarunGothic'))
)
```

```python
p = ggplot(aes(x='Attack', y='Defense'), pokemon)
```

```python
p + geom_point()
```

```python
p + geom_point(aes(color='factor(HP)'))
```

```python
p + geom_point(aes(color='Legendary'))
```

```python
p + geom_point(aes(color='Speed'))
```

```python
p + geom_point(aes(color='HP')) + scale_color_gradient(low='blue', high='red')
```

```python
p + geom_point(aes(size='HP'))
```

```python
(ggplot(pokemon)
 + geom_col(aes('Type 1', 'Attack', fill='Legendary'))
)
```

```python
(ggplot(pokemon) 
 + aes(x='Type 1', y='HP',  fill='Legendary')
 + geom_col()
 + theme(text=element_text(family='NanumBarunGothic'),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon, 
        aes(x='Type 1', y='Attack', fill='Legendary'))
 + geom_bar(stat='identity', position='dodge')
 + theme(text=element_text(family='NanumBarunGothic'),
        axis_text_x=element_text(rotation=70))
)
```

```python
pokemon = pokemon.rename(index=str, columns={"Type 1": "type1", "Type 2": "type2"})
pokemon.head()
```

```python
(ggplot(pokemon)
 + aes(x='type1', y='Total')
 + geom_col(fill='darkslateblue')
 + coord_flip()
 + theme(figure_size=(8, 6))
)
```

```python
(ggplot(pokemon)
 + aes('Attack', 'Defense')
 + aes(color='Legendary')
 + geom_point()
 + facet_wrap('type1')
)
```

```python
(ggplot(pokemon)
 + aes('Attack', 'Defense')
 + geom_bin2d(bins=10)
)
```

```python
(ggplot(pokemon)
 + aes('HP', 'type1')
 + geom_bin2d(bins=20)
)
```

```python
(ggplot(pokemon)
 + aes('type1', 'Attack')
 + geom_boxplot() 
 + theme(text=element_text(family='NanumBarunGothic'),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
 + aes('type1', 'Attack')
 + aes(color='Legendary')
 + geom_boxplot() 
 + theme(text=element_text(family='NanumBarunGothic'),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon)
 + aes('type1', 'Defense')
 + geom_boxplot() 
 + theme(text=element_text(family='NanumBarunGothic'),
        axis_text_x=element_text(rotation=70))
)
```

```python
(ggplot(pokemon, aes(x='type1', y='HP', fill='type2'))
 + geom_col()
 + theme(text=element_text(family='NanumBarunGothic'),
        axis_text_x=element_text(rotation=70))
)
```
