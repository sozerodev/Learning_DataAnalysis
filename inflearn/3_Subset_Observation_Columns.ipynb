## df.filter(regex='regex')
- 정규표현식을 의미한다
- 'regex'에서 '^a' 는 a로 시작하는 컬럼을, 'a&'는 a로 끝나는 컬럼을 가져오는 것을 의미한다. 
- 'regex'에서 '^(?!species$).*'는 특정 문자로 시작하는 것을 빼고 가져온다는 것을 의미한다.

```python
# regex = str 일때, str에 들어가는 문자를 갖고 있는 컬럼만을 가져온다.

# '_'가 들어가는 컬럼을 가져와보자.
df.filter(regex='_')
```

```python
# lenght로 끝나는 컬럼을 가져와보자.
df.filter(regex='length$')
```

```python
# 특정 문자로 시작하는 것을 가져와보고싶다면?
df.filter(regex='^sepal')
```

```python
df.filter(regex='^(?!species$).*')
```

## df.loc[:,'x2':'x4']
- 맨앞의 ':'는 모든 행을 가져온다는 의미이다. 
- 특정 컬럼 x2~x4컬럼까지 가져오라는 의미
- 따라서 df.loc[a:b, 'x2':'x4']는 'a행에서 b행까지 모두 포함하여' + 'x2~x4컬럼'을 가져온다는 것을 의미함

```python
# 두번째 컬럼인 'sepal_width'부터 네번째 컬럼인 'petal_width'컬럼까지 가져와보자
df.loc[2:5,'sepal_width' : 'petal_width']
```

## df.iloc[:, [1, 2, 5]]
- (0열부터시작한다 할때) 1, 2, 5열을 가져온다는 것을 의미
- df.iloc[a:b, [c, d, e]]는 a에서 b-1인덱스의 행까지만 가져오고, c,d,e열을 가져온다는 것을 의미함

```python
df.head()
```

```python
df.iloc[1:5, [1, 2, 3]]
```

```python
# 뒤에서 다섯개의 행을 보고싶다면?
df.iloc[-5:, [1, 3]]
```

## df.loc[df['a'] > 10, ['a','c']]
- a컬럼이 10보다 크고, a,c컬럼을 가져온다는 것.
- 앞이 행이고 뒤에가 열을 의미함.

```python
df.loc[df['sepal_length']>5, ['sepal_length', 'sepal_width']]
```
