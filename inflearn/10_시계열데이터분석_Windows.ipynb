# 시계열 데이터 분석 Windows
## 시계열데이터(Time Series Data)분석을 위한 판다스 Expanding and Rolling 이해하기
- df.expanding()
- df.rolling(n)

* https://pandas.pydata.org/pandas-docs/stable/user_guide/computation.html
* https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.expanding.html
* https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.rolling.html

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# 랜덤한 수를 2000년 1월 1일까지 생성
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2015', periods=1000))
s
```

```python
s.plot()
```

```python
# s에 누적합을 다시 더해준다.
s = s.cumsum()
# 그것을 그려보자. s의 랜덤값들을 다 더하여 누적된 것을 그려보자.
s.plot()
```

```python
# rolling은 보통 이동평균을 구할때 사용한다.
r = s.rolling(window = 60)
r
```

```python
r.mean()
```

```python
# 이동평균(moving average)
s.plot(style = 'k--')
r.mean().plot(style = 'k')
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=pd.date_range('1/1/2015', periods=1000),
                  columns=['A', 'B', 'C', 'D'])
```

```python
df
```

```python
df = df.cumsum()
df
```

```python
df.plot()
```

```python
df.rolling(window = 60).sum().plot(subplots = True)
```

```python
df.rolling(window = len(df), min_periods = 1).mean()[:5]
```

```python
df.rolling(window = len(df), min_periods = 1).mean().plot()
```

```python
df.expanding(min_periods = 1).mean()
```

```python
df.expanding(min_periods = 1).mean().plot()
```

```python
# df.expanding?에서 가져온 예제
dfe = pd.DataFrame({'B': [0, 1, 2, np.nan, 4]})
dfe
```

```python
# 결측치가 끊겨서 그려진다.
dfe.plot()
```

```python
dfe.expanding(2).sum()
```

```python
dfe.expanding(2).sum().plot()
```

```python
dfe.expanding(2).mean().plot()
```

```python
dfe.expanding(2).max().plot()
```
