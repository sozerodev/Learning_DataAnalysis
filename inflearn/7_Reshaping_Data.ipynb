## melt, pivot으로 Tidy Data만들기
- df를 어떻게 변형하냐에 따라서 필요에 따라 변형해 사용해야할 때가 있다. 
- 예를 들어 시각화, 머신러닝 등을 위해 데이터 전처리부분에서 꼭 필요하다.
<br><br>
`melt`를 통해 열로 되어있는 데이터를 행으로, <br>
`pivot`을 통해 행으로 된 데이터를 열로 보낼 수 있다.

````python
# df 새로 만들기
pd.melt?
````

````python
# pd.melt?에 있는 df가져오기
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})

df
````

### pd.melt(df)
- 컬럼의 값들을 로우로 바꿔준다

````python
# id_vars, value_vars를 통해 A는 id, B, C는 value로.

pd.melt(df, id_vars=['A'], value_vars=['B', 'C'])
````

````python
pd.melt(df, id_vars=['A'], value_vars=['B'])
````

````python
pd.melt(df, value_vars=['A','B', 'C'])
````

````python
# 열값에 D가 없으니 에러가 뜬다.
pd.melt(df, id_vars=['A'], value_vars=['D'])
````

````python
# A, B, C각각의 컬럼을 로우로 옮겼다. 
pd.melt(df, value_vars=['A', 'B', 'C'])
````

## 위의 variable, value라 되어있는 컬럼명을 바꿔줘보자.
### pandas cheat sheet의 `Method Chaining` 참고

````python
# variable은 var로, vlaue는 val로 변경하기
pd.melt(df, value_vars=['A', 'B', 'C']).rename(columns={
                   'variable' : 'var',
                   'value' : 'val'})
````

### df.pivot(columns='var', values='val')
- row로 보냈던 컬럼값들을 다시 컬럼으로 보내보자

````python
df.pivot?
````

````python
df2 = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two','two'],
                    'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                    'baz': [1, 2, 3, 4, 5, 6]})

df2
````

````python
# pivot을 통해 A, B, C라 되어있는 컬럼을 만들어서 옮겨주었다.
# index는 foo, column은 bar, 값은 baz
df2.pivot(index = 'foo', columns = 'bar', values = 'baz')
````

### df.reset_index( )
- A, B, C라는 컬럼이 만들어진 후 df.reset_index()를 하면 왼쪽에 bar와 foo값이 각각의 컬럼명으로 들어가있는 것을 볼 수 있다.
- 즉, 새로운 index를 만들어준 것

````python
df2.pivot(index = 'foo', columns = 'bar', values = 'baz').reset_index()
````

````python
# 이를 df3에 담아줘보자.
df3 = df2.pivot(index = 'foo', columns = 'bar', values = 'baz').reset_index()
df3
````

````python
# pd.melt()를 다시해보자.
# pivot으로 만든 데이터를 melt를 통해.. df2라 만들어준 데이터프레임과 유사하게 다시 만들어보자.
df3.melt?

# 여기의 자료를 가져와보자
````

````python
df3.melt(id_vars=['foo'], value_vars=['A', 'B', 'C'])
````

````python
# 위의 데이터프레임을 df2와 유사하게 만들기 위해
# df.sort_values()를 통해 method chaining형태로 만들 수 있다.
# 'foo'를 통해 정리해보자
df3.melt(id_vars=['foo'], value_vars=['A', 'B', 'C']).sort_values('foo')
````

````python
# 위와 같은 정렬형태를 `리스트`를 통해 두개의 열로 정리할 수도 있다.
df3.melt(id_vars=['foo'], value_vars=['A', 'B', 'C']).sort_values(['foo', 'bar'])
````

````python
# method chaining을 좀더 응용해보자
# 컬럼의 이름도 바꿔줘보자. value를 baz로 바꿔줘보자
df3.melt(
    id_vars=['foo'],
    value_vars=['A', 'B', 'C']).sort_values(
    ['foo', 'bar']).rename(columns = {'value':'baz'})
````

## pd.concat([df1, df2]) 시리즈, 데이터프레임 합치기
- 서로 다른 데이이터인 df1, df2를 하나의 데이터프레임으로 합쳐주기.
- 같은 컬럼이 있는 두 개의 데이터를 `위아래`로 합쳐주기
- 리스트 형태로 두개를 합쳐주는 것을 보여줬지만 리스트를 통해 여러개의 데이터를 합치는 것도 가능하다.
<br><br>
- 외부데이터를 가져와 합치거나 전처리한 데이터를 합치거나 할때 pd.concat()을 많이 사용한다. 

````python
pd.concat?
````

````python
# pd.concat?의 공식문서에서 예제를 가져옴
s1 = pd.Series(['a', 'b'])
s1
````

````python
s2 = pd.Series(['c', 'd'])
s2
````

````python
# pd.concat()을 통해 s1과 s2를 합쳐주자
pd.concat([s1, s2])
````

````python
# 이때 pd.concat()에서 'ignore_index = True'라는 옵션을 줄 수 있다.
# 0 1 0 1로 되어있던 인덱스가 0 1 2 3으로 바뀐 것을 볼 수 있다.
pd.concat([s1, s2], ignore_index = True)
````

````python
# s1, s2로 키를 줘보기
pd.concat([s1, s2], keys = ['s1', 's2'])
````

````python
# names를 통해 컬럼명 지정도 가능하다.
pd.concat([s1, s2], keys = ['s1', 's2'], names = ['Series name', 'Row ID'])
````

### pd.concat?에서 df형태를 가져와 새로운 실습을 해보자

````python
df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['letter', 'number'])
df1
````

````python
df2 = pd.DataFrame([['c', 3], ['d', 4]],
                   columns=['letter', 'number'])
df2
````

````python
# concat()을 통해 df1, df2를 합쳐줘보자
pd.concat([df1, df2])
````

````python
# df3라는 데이터프레임을 또 만들어보자
df3 = pd.DataFrame([['c', 3, 'cat'], ['d', 4, 'dog']],
                   columns=['letter', 'number', 'animal'])
df3
````

````python
# df1, df3를 합쳐보자.
# 형태가 다른 데이터프레임들을 합쳐보자
pd.concat([df1, df3])
````

### concat()을 할때 `join = "inner"`라는 옵션을 줘보자

````python
pd.concat([df1, df3], join = "inner")
````

=> animal이라는 컬럼이 없어지고 letter, number만 조인된 것을 확인 가능하다.

### 이때, `verify_integrity = True`라는 옵션을 주게되면?
### 에러가 나게 된다.

````python
pd.concat([df5, df6], verify_integrity = True)
````

ValueError가 난다.<br>
왜냐하면 같은 index값이 있기에...<br>
