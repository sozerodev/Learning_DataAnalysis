## pandas로 apply활용하기, lambda 익명함수 사용하기

### apply(function)
- 데이터 전처리에 있어서 유용하다.

````python
df.apply?
````

````python
# 익명함수 lambda
# 데이터를 처리할때 이러한 익명함수 lambda를 사용하기도 한다.
df.apply(lambda x : x)
````

````python
print(df.apply(lambda x : x[0]), end="\n\n\n")
print(df.apply(lambda x : x[1]), end="\n\n\n")
print(df.apply(lambda x : x[2]))
````

````python
# species컬럼에서 0번째항..세번째 글자까지 가져올 수 있다.
df['species'].apply(lambda x : x[:3]).head()
````

````python
# 따로 컬럼을 만들어 값을 넣어주는 것도 가능하다. 
df['species_3'] = df['species'].apply(lambda x : x[:3])
df
````

````python
# 함수를 아예 새로만들어보자
def smp(x) : # sample의 약자 smp
    # 뒤에서 3번째 까지의 문자를 가져오는 함수
    x = x[:-3] # 뒤에서 세번째 값을 가져오자
    return x
````

````python
df['species-3'] = df['species'].apply(smp)
df
````

## 따라서 apply()함수를 사용할때
- 1. lambada 다음에 익명함수를 사용해도 되고

```python
df['species_3'] = df['species'].apply(lambda x : x[:3])
```

- 2. 그 전에 외부에다가 함수를 하나 만들어서 호출하여 사용해도 된다.

```python
def smp(x) :
    # 뒤에서 3번째 까지의 문자를 가져오는 함수
    x = x[:-3]
    return x

df['species-3'] = df['species'].apply(smp)
```
